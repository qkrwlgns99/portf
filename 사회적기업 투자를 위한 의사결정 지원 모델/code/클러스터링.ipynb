#경제적 성과 PCA


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# '경제적 성과' 열을 제외한 특성 데이터
features = support_f.drop(['경제적 성과'], axis=1)

# 특성 데이터를 표준화
scaler = StandardScaler()
selected_features_scaled = scaler.fit_transform(features)

# PCA 차원 축소.
pca = PCA(n_components=2)
selected_features_pca = pca.fit_transform(selected_features_scaled)

# PCA로 차원 축소된 데이터를 DataFrame으로 변환하여 'P1'과 'P2'라는 열 이름을 부여
selected_features_pca_df = pd.DataFrame(selected_features_pca, columns=['P1', 'P2'])

# PCA로 차원 축소된 데이터와 '경제적 성과' 열을 합친 최종 데이터프레임
final_df = pd.concat([selected_features_pca_df, support_f['경제적 성과']], axis=1)

# 결과 출력
print(final_df.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 원본 데이터에서 '경제적 성과' 열을 제외한 수치형 특성 데이터
numeric_features = support_f.drop(['경제적 성과'], axis=1)

# 표준화 (Z-score 정규화)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(numeric_features)

# K-Means 알고리즘으로 클러스터링 수행
kmeans = KMeans(n_clusters=5)
kmeans.fit(scaled_features)

# 각 데이터 포인트의 클러스터 할당
clusters = kmeans.predict(scaled_features)

# 클러스터별 평균 계산
cluster_means = pd.DataFrame(scaled_features).groupby(clusters).mean()

# 전체 데이터의 평균 계산
overall_mean = scaled_features.mean().mean()

# 시각화
plt.figure(figsize=(20, 6))
plt.plot(cluster_means.T, marker='o', label='클러스터 평균값')
plt.axhline(y=overall_mean, color='r', linestyle='dashed', label='Overall Mean')
plt.xlabel('Features')
plt.ylabel('Mean Value (정규화)')
plt.title('Cluster Mean vs. Overall Mean for Each Feature (정규화)')
plt.xticks(range(len(numeric_features.columns)), numeric_features.columns, rotation=90)
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 원본 데이터에서 '사회적 성과' 열을 제외한 수치형 특성 데이터
s_numeric_features = support_s.drop(['사회적 성과'], axis=1)

# 표준화 (Z-score 정규화)
scaler = StandardScaler()
s_scaled_features = scaler.fit_transform(s_numeric_features)

# K-Means 알고리즘으로 클러스터링 수행
kmeans = KMeans(n_clusters=6)
kmeans.fit(s_scaled_features)

# 각 데이터 포인트의 클러스터 할당
clusters = kmeans.predict(s_scaled_features)

# 클러스터별 평균 계산
cluster_means = pd.DataFrame(s_scaled_features).groupby(clusters).mean()

# 전체 데이터의 평균 계산
overall_mean = s_scaled_features.mean().mean()

# 시각화
plt.figure(figsize=(20, 6))
plt.plot(cluster_means.T, marker='o', label='클러스터 평균값')
plt.axhline(y=overall_mean, color='r', linestyle='dashed', label='Overall Mean')
plt.xlabel('Features')
plt.ylabel('Mean Value (정규화)')
plt.title('Cluster Mean vs. Overall Mean for Each Feature (정규화)')
plt.xticks(range(len(numeric_features.columns)), numeric_features.columns, rotation=90)
plt.legend()
plt.show()
```
